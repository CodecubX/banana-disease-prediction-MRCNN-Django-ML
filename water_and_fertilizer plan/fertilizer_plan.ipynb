- **pH of soil (measured on a scale of 0-14)**: Soil pH can affect the availability of nutrients
to plants. For example, in acidic soils (pH < 6.0), certain nutrients like Phosphorus and
Potassium tend to be less available to plants.
Organic matter content in soil (low, moderate, high): Soils with higher organic matter
content tend to have higher levels of Nitrogen, Phosphorus, and Potassium, which can
contribute to better plant growth and nutrient availability.

- **Soil type ('Black Soil', 'Cinder Soil', 'Laterite Soil', 'Peat Soil', 'Yellow Soil')**: The
type of soil can affect its water-holding capacity, nutrient-holding capacity, and drainage,
which in turn can influence the watering requirements for plants.

- **Soil moisture (measured as a percentage)**: Soil moisture can affect the availability of
nutrients to plants as well as plant growth and yield. It indicates the amount of water
present in the soil at a given time.

- **Average temperature (measured in Celsius)**: Temperature can affect plant growth and
development as well as the rate of water evaporation from the soil. It is an important
factor to consider when determining the watering schedule.

- **Average rainfall (measured in millimeters per month)**: Rainfall can impact the
watering needs of plants as it contributes to the overall soil moisture content. Higher
rainfall may require less frequent watering, while lower rainfall may necessitate more
frequent watering.

- **Plant height (measured in centimeters)**: Plant height can be an indicator of plant
growth and development. Taller plants may have higher water requirements compared to
shorter ones.

- **Leaf color (Green, Yellow, Brown)**: Leaf color can indicate the overall health and water
status of the plant. Yellow or brown leaves may be a sign of water stress or inadequate
watering.

- **Stem diameter (measured in millimeters)**: Stem diameter can be an indicator of plant
growth and development. Thicker stems may require more water compared to thinner
ones.

- **Plant density (number of plants per unit area)**: Plant density can affect water uptake
and water competition among plants. Higher plant density may require more frequent
watering to ensure all plants receive sufficient water.

- **Soil texture**: Soil texture can give an indication of the water-holding capacity of the soil.
Clayey soils tend to hold water for longer periods compared to sandy soils, influencing
the frequency and amount of watering required.

- **Soil color**: The color of the soil can provide insights into its composition and
water-retention capabilities. Dark-colored soils tend to have higher organic matter
content and better water retention capacity.

- **Temperature (in degrees Celsius)**: Temperature affects evaporation rates and can
impact the rate at which water is lost from the soil. Higher temperatures may lead to
increased evaporation and higher watering needs.

- **Humidity (in percentage)**: Humidity levels in the air can influence the rate of water loss
from the soil and the plant. Lower humidity levels may increase the rate of evaporation
and result in higher watering requirements.

- **Rainfall (in mm)**: Rainfall directly contributes to the soil moisture content and can impact
the watering needs of plants. Higher rainfall may reduce the frequency of additional
watering.

- **Water source ('Well', 'River', 'Canal', 'Rainwater harvesting', 'Municipal supply')**:
The source of water used for irrigation can affect the quality and availability of water for
plants.

- **Irrigation method ('Drip', 'Flood', 'Sprinkler')**: The method of irrigation used can
impact the efficiency of water delivery to the plants and influence the watering schedule.

- **Fertilizer type used in the previous season ('Organic', 'Inorganic', 'both')**: The type
of fertilizer used in the previous season can affect soil nutrient availability and may
impact the watering requirements for the current season.

- **Crop rotation ('Yes', 'No')**: Crop rotation can help to prevent nutrient depletion and
reduce pest and disease incidence, which may indirectly impact the watering needs of
plants.

- **Pest and disease infestation ('Yes', 'No')**: Pest and disease infestation can affect plant
growth, yield, and water requirements. Infested plants may require additional watering to
cope with stress.

- **Slope of the land ('None', 'Low', 'Medium', 'High')**: The slope of the land can affect
water drainage and runoff, which can influence the watering requirements for plants.

##### Target Variables
- **Fertilizer type** (organic, inorganic): The type of fertilizer recommended for the current
season based on soil nutrient content and plant requirements.
- **Fertilizer plan** (dose and timing): The recommended dose and timing for applying
fertilizer based on soil nutrient content, plant requirements, and other factors such as
irrigation method and crop rotation.

# Encoding

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
import pickle

SAVED_DIR = './fertilizer_plan_saved_data/'

def encode_categorical_features(df, one_hot_cols_fertilizer_type, one_hot_cols_fertilizer_plan, ordinal_cols, target_column_1, target_column_2):
    # One-hot encode categorical columns -fertilizer type
    one_hot_encoder_fertilizer_type = OneHotEncoder(sparse=False, drop='first')
    one_hot_encoded_fertilizer_type_data = one_hot_encoder_fertilizer_type.fit_transform(df[one_hot_cols_fertilizer_type])

    # Save the encoder object using pickle
    one_hot_encoder_filename = f'{SAVED_DIR}one_hot_encoder_fertilizer_type.pkl'
    with open(one_hot_encoder_filename, 'wb') as f:
        pickle.dump(one_hot_encoder_fertilizer_type, f)

    # One-hot encode categorical columns -fertilizer type
    one_hot_encoder_fertilizer_plan = OneHotEncoder(sparse=False, drop='first')
    one_hot_encoded_fertilizer_plan_data = one_hot_encoder_fertilizer_plan.fit_transform(df[one_hot_cols_fertilizer_plan])

    # Save the encoder object using pickle
    one_hot_encoder_filename = f'{SAVED_DIR}one_hot_encoder_fertilizer_plan.pkl'
    with open(one_hot_encoder_filename, 'wb') as f:
        pickle.dump(one_hot_encoder_fertilizer_plan, f)

    # Ordinal encode categorical columns: uses one ordinal encoder for fertilizer type and plan
    ordinal_encoder = OrdinalEncoder()
    ordinal_encoded_data = ordinal_encoder.fit_transform(df[ordinal_cols])

    # Save the encoder object using pickle
    ordinal_encoder_filename = f'{SAVED_DIR}ordinal_encoder.pkl'
    with open(ordinal_encoder_filename, 'wb') as f:
        pickle.dump(ordinal_encoder, f)

    # Target variable 1 - fertilizer_type
    ordinal_encoder_target_var_1 = OrdinalEncoder()
    ordinal_encoder_target_var_1_data = ordinal_encoder_target_var_1.fit_transform(df[target_column_1])

    # Save the encoder object using pickle - target_var 1
    ordinal_encoder_target_var_1_filename = f'{SAVED_DIR}ordinal_encoder_target_var_fertilizer_type.pkl'
    with open(ordinal_encoder_target_var_1_filename, 'wb') as f:
       pickle.dump(ordinal_encoder_target_var_1, f)

    # Target variable 2 - fertilizer_plan
    ordinal_encoder_target_var_2 = OrdinalEncoder()
    ordinal_encoder_target_var_2_data = ordinal_encoder_target_var_2.fit_transform(df[target_column_2])

    # Save the encoder object using pickle
    ordinal_encoder_target_var_2_filename = f'{SAVED_DIR}ordinal_encoder_target_var_fertilizer_plan.pkl'
    with open(ordinal_encoder_target_var_2_filename, 'wb') as f:
       pickle.dump(ordinal_encoder_target_var_2, f)

    return one_hot_encoded_fertilizer_type_data, one_hot_encoded_fertilizer_plan_data,  ordinal_encoded_data, ordinal_encoder_target_var_1_data, ordinal_encoder_target_var_2_data
```

```python
df.columns
```

```python
# Ordinal columns (ones that have an order)
ordinal_columns = ['organic_matter_content', 'slope']
# Categorical columns (ones without order, already encoded)
categorical_columns_fertilizer_type = ['soil_type', 'leaf_color', 'soil_texture',
                       'soil_color', 'water_source', 'irrigation_method',
                       'fertilizer_used_last_season', 'crop_rotation',
                       'pest_disease_infestation']

categorical_columns_fertilizer_plan = categorical_columns_fertilizer_type.copy()
categorical_columns_fertilizer_plan.append('fertilizer_type')

# Non-categorical columns
non_categorical_columns = ['pH', 'soil_moisture', 'avg_temperature', 'avg_rainfall',
                           'plant_height', 'stem_diameter', 'plant_density',
                           'temperature', 'humidity', 'rainfall']
target_column_1 = ['fertilizer_type']
target_column_2 = ['fertilizer_plan']
```

```python
# encoding data
one_hot_encoded_fertilizer_type_data, one_hot_encoded_fertilizer_plan_data, ordinal_encoded_data,\
ordinal_encoded_fertilizer_type_data, \
ordinal_encoded_fertilizer_plan_data = encode_categorical_features(
    df, categorical_columns_fertilizer_type, categorical_columns_fertilizer_plan, ordinal_columns, target_column_1, target_column_2)
```

```python
# get numerical data columns into a ndarray
numerical_columns = df[non_categorical_columns].values
```

```python
import numpy as np

# create independant variables
x_fertilizer_type = np.hstack([
    one_hot_encoded_fertilizer_type_data,
    ordinal_encoded_data,
    numerical_columns
])
x_fertilizer_plan = np.hstack([
    one_hot_encoded_fertilizer_plan_data,
    ordinal_encoded_data,
    numerical_columns
])
y_fertilizer_type = ordinal_encoded_fertilizer_type_data
y_fertilizer_plan = ordinal_encoded_fertilizer_plan_data
```

```python
##### shapes
print('OHE shape: ',one_hot_encoded_fertilizer_type_data.shape, '\nOE shape: ', ordinal_encoded_data.shape)
print('OHE shape: ',one_hot_encoded_fertilizer_plan_data.shape, '\nOE shape: ', ordinal_encoded_data.shape)
print('x fertilizer type shape:', x_fertilizer_type.shape)
print('x fertilizer plan shape:', x_fertilizer_plan.shape)
print('y fertilizer type shape: ', y_fertilizer_type.shape)
print('y fertilizer plan shape: ', y_fertilizer_plan.shape)
```

# Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler_fertilizer_type = StandardScaler()
x_fertilizer_type = scaler_fertilizer_type.fit_transform(x_fertilizer_type)

with open(f'{SAVED_DIR}fertilizer_type_scaler.pkl', 'wb') as f:
    pickle.dump(scaler_fertilizer_type, f)

scaler_fertilizer_plan = StandardScaler()
x_fertilizer_plan = scaler_fertilizer_plan.fit_transform(x_fertilizer_plan)

with open(f'{SAVED_DIR}fertilizer_plan_scaler.pkl', 'wb') as f:
    pickle.dump(scaler_fertilizer_plan, f)
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split

# Split the data into train and test sets
x_train_fertilizer_type, x_test_fertilizer_type, y_train_fertilizer_type, y_test_fertilizer_type = train_test_split(x_fertilizer_type, y_fertilizer_type, test_size=0.2, random_state=42, stratify=y_fertilizer_type)
x_train_fertilizer_plan, x_test_fertilizer_plan, y_train_fertilizer_plan, y_test_fertilizer_plan = train_test_split(x_fertilizer_plan, y_fertilizer_plan, test_size=0.2, random_state=42, stratify=y_fertilizer_plan)
```

# Model Building
> `OneVsRestClassifier` approach, which trains multiple binary classifiers, one for each class, and then combines their predictions.

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc
```

> The roc_curve function does not support multiclass classification directly.
>
> However, can adapt the code to compute the **ROC curve** and **AUC score** for
> each class separately using the **One-vs-Rest (OvR)** strategy.

```python
# Convert the target variable to binary form
y_train_fertilizer_type_bin = label_binarize(y_train_fertilizer_type, classes=np.unique(y_train_fertilizer_type))
y_test_fertilizer_type_bin = label_binarize(y_test_fertilizer_type, classes=np.unique(y_test_fertilizer_type))

print(y_train_fertilizer_type_bin.shape, y_test_fertilizer_type_bin.shape)

y_train_fertilizer_plan_bin = label_binarize(y_train_fertilizer_plan, classes=np.unique(y_train_fertilizer_plan))
y_test_fertilizer_plan_bin = label_binarize(y_test_fertilizer_plan, classes=np.unique(y_test_fertilizer_plan))

print(y_train_fertilizer_plan_bin.shape, y_test_fertilizer_plan_bin.shape)
```

### Random Forest Classifier

#### Fertilizer Type

```python
rf = OneVsRestClassifier(RandomForestClassifier())
rf.fit(x_train_fertilizer_type, y_train_fertilizer_type)
rf_probs_fertilizer_type = rf.predict_proba(x_test_fertilizer_type)
print(y_test_fertilizer_type_bin.shape, rf_probs_fertilizer_type.shape)
rf_fpr, rf_tpr, rf_thresholds = roc_curve(y_test_fertilizer_type_bin.ravel(), rf_probs_fertilizer_type.ravel())
rf_auc_fertilizer_type = auc(rf_fpr, rf_tpr)
```

#### Fertilizer Plan

```python
rf = OneVsRestClassifier(RandomForestClassifier())
rf.fit(x_train_fertilizer_plan, y_train_fertilizer_plan)
rf_probs_fertilizer_plan = rf.predict_proba(x_test_fertilizer_plan)
print(y_test_fertilizer_plan_bin.shape, rf_probs_fertilizer_plan.shape)
rf_fpr, rf_tpr, rf_thresholds = roc_curve(y_test_fertilizer_plan_bin.ravel(), rf_probs_fertilizer_plan.ravel())
rf_auc_fertilizer_plan = auc(rf_fpr, rf_tpr)
```

## Linear Regression

#### Fertilizer Type

```python
lr = OneVsRestClassifier(LogisticRegression(max_iter=1000))
lr.fit(x_train_fertilizer_type, y_train_fertilizer_type)
lr_probs = lr.predict_proba(x_test_fertilizer_type)
lr_fpr, lr_tpr, lr_thresholds = roc_curve(y_test_fertilizer_type_bin.ravel(), lr_probs.ravel())
lr_auc_fertilizer_type = auc(lr_fpr, lr_tpr)
```

#### Fertilizer Plan

```python
lr = OneVsRestClassifier(LogisticRegression(max_iter=1000))
lr.fit(x_train_fertilizer_plan, y_train_fertilizer_plan)
lr_probs = lr.predict_proba(x_test_fertilizer_plan)
lr_fpr, lr_tpr, lr_thresholds = roc_curve(y_test_fertilizer_plan_bin.ravel(), lr_probs.ravel())
lr_auc_fertilizer_plan = auc(lr_fpr, lr_tpr)
```

## Gradient Boosting Classifier

#### Fertilizer Type

```python
gb = OneVsRestClassifier(GradientBoostingClassifier())
gb.fit(x_train_fertilizer_type, y_train_fertilizer_type)
gb_probs = gb.predict_proba(x_test_fertilizer_type)
gb_fpr, gb_tpr, gb_thresholds = roc_curve(y_test_fertilizer_type_bin.ravel(), gb_probs.ravel())
gb_auc_fertilizer_type = auc(gb_fpr, gb_tpr)
```

#### Fertilizer Plan

```python
gb = OneVsRestClassifier(GradientBoostingClassifier())
gb.fit(x_train_fertilizer_plan, y_train_fertilizer_plan)
gb_probs = gb.predict_proba(x_test_fertilizer_plan)
gb_fpr, gb_tpr, gb_thresholds = roc_curve(y_test_fertilizer_plan_bin.ravel(), gb_probs.ravel())
gb_auc_fertilizer_plan = auc(gb_fpr, gb_tpr)
```

## XGBoost Classifier

#### Fertilizer Type

```python
xgb = OneVsRestClassifier(XGBClassifier())
xgb.fit(x_train_fertilizer_type, y_train_fertilizer_type)
xgb_probs = xgb.predict_proba(x_test_fertilizer_type)
xgb_fpr, xgb_tpr, xgb_thresholds = roc_curve(y_test_fertilizer_type_bin.ravel(), xgb_probs.ravel())
xgb_auc_fertilizer_type = auc(xgb_fpr, xgb_tpr)
```

#### Fertilizer Plan

```python
xgb = OneVsRestClassifier(XGBClassifier())
xgb.fit(x_train_fertilizer_plan, y_train_fertilizer_plan)
xgb_probs = xgb.predict_proba(x_test_fertilizer_plan)
xgb_fpr, xgb_tpr, xgb_thresholds = roc_curve(y_test_fertilizer_plan_bin.ravel(), xgb_probs.ravel())
xgb_auc_fertilizer_plan = auc(xgb_fpr, xgb_tpr)
```

```python
lgb = OneVsRestClassifier(LGBMClassifier())
lgb.fit(x_train_fertilizer_type, y_train_fertilizer_type)
lgb_probs = lgb.predict_proba(x_test_fertilizer_type)
lgb_fpr, lgb_tpr, lgb_thresholds = roc_curve(y_test_fertilizer_type_bin.ravel(), lgb_probs.ravel())
lgb_auc_fertilizer_type = auc(lgb_fpr, lgb_tpr)
```

```python
import pickle
import numpy as np

def predict(data, target='fertilizer_type', model_file='fertilizer_type_classifier.pkl', ordinal_encoder_file='ordinal_encoder.pkl', one_hot_encoder_file='one_hot_encoder_fertilizer_type.pkl', scaler_file="fertilizer_type_scaler.pkl", ordinal_encoded_target_file='ordinal_encoder_target_var_fertilizer_type.pkl'):
    """
    Performs prediction using a trained model on the given data.

    Parameters:
        data (dict): A dictionary containing the input data for prediction.
        target (str): The target variable for prediction. Default is 'fertilizer_type'.
        model_file (str): The file name of the trained model. Default is 'fertilizer_type_classifier.pkl'.
        ordinal_encoder_file (str): The file name of the ordinal encoder. Default is 'ordinal_encoder.pkl'.
        one_hot_encoder_file (str): The file name of the one-hot encoder. Default is 'one_hot_encoder_fertilizer_type.pkl'.
        scaler_file (str): The file name of the scaler. Default is 'fertilizer_type_scaler.pkl'.
        ordinal_encoded_target_file (str): The file name of the ordinal encoded target variable. Default is 'ordinal_encoder_target_var_fertilizer_type.pkl'.

    Returns:
        tuple: A tuple containing the top prediction and a dictionary of class names with corresponding probabilities.
            - top_prediction (str): The top predicted class label.
            - result (dict): A dictionary mapping class names to their corresponding probabilities.

    Note:
        - This function assumes that the required files for the model, encoders, and scaler are stored in a specific directory.
          Make sure to provide the correct file names and paths when calling this function.

    Example usage:
        sample_data = {
            'pH': 6.2,
            'organic_matter_content': 'low',
            ...
        }
        configs = {
            'target': 'fertilizer_plan',
            'model_file': 'fertilizer_plan_classifier.pkl',
            'ordinal_encoder_file': 'ordinal_encoder.pkl',
            'one_hot_encoder_file': 'one_hot_encoder_fertilizer_plan.pkl',
            'scaler_file': "fertilizer_plan_scaler.pkl",
            'ordinal_encoded_target_file': 'ordinal_encoder_target_var_fertilizer_plan.pkl'
        }
        fertilizer_type_pred, confidence = predict(sample_data.-, **configs)
    """

    # Ordinal columns (ones that have an order)
    ordinal_columns = ['organic_matter_content', 'slope']
    # Categorical columns (ones without order, already encoded)
    categorical_columns = ['soil_type', 'leaf_color', 'soil_texture',
                           'soil_color', 'water_source', 'irrigation_method',
                           'fertilizer_used_last_season', 'crop_rotation',
                           'pest_disease_infestation']
    if target == 'fertilizer_plan':
        categorical_columns.append('fertilizer_type')

    # Non-categorical columns
    non_categorical_columns = ['pH', 'soil_moisture', 'avg_temperature', 'avg_rainfall',
                               'plant_height', 'stem_diameter', 'plant_density',
                               'temperature', 'humidity', 'rainfall']

    df = pd.DataFrame.from_dict([data])

    # Load the one-hot encoder
    with open(f'{SAVED_DIR}{one_hot_encoder_file}', 'rb') as f:
        one_hot_encoder = pickle.load(f)

    # Load the ordinal encoder
    with open(f'{SAVED_DIR}{ordinal_encoder_file}', 'rb') as f:
        ordinal_encoder = pickle.load(f)

    one_hot_encoded_data = one_hot_encoder.transform(df[categorical_columns])
    ordinal_encoded_data = ordinal_encoder.transform(df[ordinal_columns])

    # Get numerical data columns into a ndarray
    numerical_columns = df[non_categorical_columns].values

    # Create independent variables
    x = np.hstack([
        one_hot_encoded_data,
        ordinal_encoded_data,
        numerical_columns
    ])

    # Load the scaler
    with open(f'{SAVED_DIR}{scaler_file}', 'rb') as f:
        scaler = pickle.load(f)

    x = scaler.transform(x)

    # Load the model from the file
    with open(f'{SAVED_DIR}{model_file}', 'rb') as f:
        model = pickle.load(f)

    # Perform inference
    predictions = model.predict(x)
    probabilities = model.predict_proba(x)

    # Reverse transform the predicted labels
    with open(f'{SAVED_DIR}{ordinal_encoded_target_file}', 'rb') as f:
        ordinal_encoder_target = pickle.load(f)

    # conert [[9.97596309e-01 1.23045918e-05 2.39138657e-03]] to [[0 1 2]]
    class_indexes = np.arange(probabilities.shape[1]).reshape(1, -1)
    # to [[0] [1] [2]]
    formatted_class_indexes = np.transpose(class_indexes)

    class_names = ordinal_encoder_target.inverse_transform(formatted_class_indexes).flatten()

    # Create a dictionary of class names and probabilities
    result = {}
    for class_name, probability in zip(class_names, probabilities[0]):
        result[class_name] = format(probability, '.5f')

    top_prediction = ordinal_encoder_target.inverse_transform(predictions.reshape(1, -1)).flatten()[0]

    return top_prediction, result
```
